```python
#import pandas
import pandas as pd
```

```python
#load dataset and have a look
df = pd.read_csv('DataSF_Restaurant_Inspections.csv')
df.head()
```

```python
#more info about content
df.info()
```

```python
#there are approx. 52k records. We in Navigator are interested in displaying the name of restaurant, it's location and phone number.
#Columns regarding inspections are of no use and can be dropped as well as removing duplicates of restaurant names regarding multiple inspections.
df = df.drop(['inspection_id', 'inspection_date','inspection_score','inspection_type','violation_id','violation_description','risk_category'], axis=1)

```

```python
#Defining record quality:
#   high quality - Full rows, data verified and validated
#   medioum quality - Full rows, lack proper verification or validation
#   low quality - Rows with empty fields
# We can see so far there are 52315 rows
```

```python
#Next we remove duplicates and drop rows with missing values
df = df.drop_duplicates(subset=['business_name'])
df = df.dropna()
df.info()
```

```python
#Now we have 859 record that fall into either medium or high quality category 
#In the following cells we will proceed with validation and verification
#Business name, address and telephone would have to be checked for each record indvidually, a process which exceedes our time and resources at the moment
#First we check the city column
df['business_city'].unique()
```

```python
#We can see that all of the cities in the records are 'San Francisco' which is what we need so this column is verified and validated
#We proceed with the business state column
df['business_state'].unique()
```

```python
#Googling state coed in the USA shows us the only allowed code fore the state of California is 'CA'
#'CA' and 'IL' are valid state codes but only 'CA' is verified so we will drop 'California' and 'IL' from our search for high quality records
df = df[df.business_state.isin(['California', 'IL']) == False]
df['business_state'].unique()
```

```python
#Now on to te postal code column
df['business_postal_code'].unique()
```

```python
#Codes like '941102019', '941', 'CA', are invalid while '92672'is valid but does not represent SF zip code according to 
#https://www.usmapguide.com/california/san-francisco-zip-code-map/
df = df[df.business_postal_code.isin(['941102019','941','94158','CA', '92672']) == False]
df['business_postal_code'].unique()
```

```python
#Last thing to check is the geolocation of the restaurants
#We weill do this by putting all of the coordinates on a map to ensure they are all in SF
import folium
SF_COORDINATES = (37.76, -122.45)
map = folium.Map(location=SF_COORDINATES, zoom_start=3) 

for each in df.iterrows():
    folium.Marker([each[1]["business_latitude"],each[1]["business_longitude"]],
        popup=each[1]["business_name"]).add_to(map),
map
```

```python
#From the map above wee see that not all of the cooridinates are in SF despite google search of a random sample of 10 restaurants ouside of SF on the map clearly shows
#they indeed are in SF. Coordinates are valid but not verified and do not meet our high quality data standards.
#We filter them by dropping ones out of SF coordinates rectangle ([37.598979 , 37.841721] , [-122.596810 , -122.343742])
df = df.loc[(df.business_latitude > 37.598979) & (df.business_latitude < 37.841721) & 
(df.business_longitude > -122.596810) & (df.business_longitude < -122.343742)].reset_index(drop=True)
```

```python
SF_COORDINATES = (37.76, -122.45)
map = folium.Map(location=SF_COORDINATES, zoom_start=3) 

for each in df[0:885].iterrows():
    folium.Marker([each[1]["business_latitude"],each[1]["business_longitude"]],
        popup=each[1]["business_name"]).add_to(map).add_to(map)
map
```

```python
#Now all of our record are indeed restaurants in SF, with proper state code, zip code and location validated and verified
#Checking the amount of high quality records
df.info()
```
